```python
# Cleaning function for standardizing branch names
replacements = {
    "all (ug/pg)": "All UG/PG",
    "all(ug/pg)": "All UG/PG",
    "all (ug and pg)": "All UG/PG",
    "all (pg)": "All PG",

    "all(ug)": "All UG",
    "all (ug)": "All UG",
    "all ug": "All UG",
    "all(ug)": "All UG",
    "alL UG": "All UG",
    "ALl UG": "All UG",
    "ALL UG": "All UG",
    "ALL(UG)": "All UG",

    "ce & it (ug)": "CE & IT UG",
    "ce & i t (ug)": "CE & IT UG",
    "ce &i t (ug)": "CE & IT UG",
    "ce & it (ug/pg)": "CE & IT UG/PG",
    "ce & it (ug and pg)": "CE & IT UG/PG",
    "ce & it (ug & pg)": "CE & IT UG/PG",
    "ce & it (ug - female)": "CE & IT UG (Female)",
    "ce+it+entc(girls)": "CE & IT & ENTC (Girls)",
    "ce&it": "CE & IT UG",
    
    "comp&it (ug/pg)": "CE & IT UG/PG",
    "comp&it(ug)": "CE & IT UG",
    "comp & it ug": "CE & IT UG",
    "comp&e&tc(ug)": "CE & ENTC UG",

    "e&tc (ug)": "ENTC UG",
    "e&tc (ug/pg)": "ENTC UG/PG",
    "e&tc (ug&pg)": "ENTC UG/PG",
    "entc": "ENTC",

    "cs & it (ug)": "CE & IT UG",
    "cs&it": "CE & IT UG",
    "cs": "CE",

    "it(ug)": "IT UG",
    
    "all (ug) & ce (pg)": "All UG",
    "all (ug) and ce (pg)": "All UG",
    "all (ug) and ce & it (pg)": "All UG",
    "all (ug - female)": "All UG (Female)",
}

```

```python
df['branches'] = df['branches'].str.lower().str.strip().replace(replacements)
```

```python
df.loc[df['year'] == 2018, 'cgpa'] = df['cgpa'] / 10
```
